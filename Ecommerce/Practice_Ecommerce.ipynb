```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

# 2) Count the number of orders placed in 2017

query = '''
            SELECT COUNT(orders.order_id) from ecommerce.orders
            WHERE YEAR(orders.order_purchase_timestamp) = 2017
        '''
cur.execute(query)
data = cur.fetchall()
print(f"{data[0][0]} - orders placed in 2017")
```

```python
  # Total orders placed in Year-wise
query = '''
            SELECT COUNT(orders.order_id), YEAR(orders.order_purchase_timestamp) Year_Wise from ecommerce.orders
            GROUP BY YEAR(orders.order_purchase_timestamp)
        '''
cur.execute(query)
data = cur.fetchall()
df = pd.DataFrame(data, columns = ["Count Of Orders Placed", "Year"])
#df
colors = ["yellow", "lightGreen", "red"]
plt.pie(df['Count Of Orders Placed'], labels = df["Year"] , colors = colors, shadow = "True", startangle = -90, labeldistance=1.2, autopct='%0.1f%%') # autopct -> shows percentage
plt.title("Orders Place Year Wise")
#plt.axis('equal')  # Keeps the pie chart circular
plt.show()
```

```python
# 3) Find the total sales per category

query = """
        SELECT products.product_category, round(SUM(payments.payment_value), 2) AS Revenue
        	FROM  ecommerce.products 
             JOIN ecommerce.order_items ON order_items.product_id = products.product_id
             JOIN ecommerce.payments ON payments.order_id = order_items.order_id
        GROUP BY products.product_category
        ORDER BY  product_category
        """
cur.execute(query)
data = cur.fetchall()
df = pd.DataFrame(data, columns = ["Product_Category", "Revenue"])
df = df.sort_values(by = "Product_Category", ascending=False)

# Error you're seeing is caused by the presence of None values (i.e. missing data) in your Product_Category column. Matplotlib expects string labels for the x-axis in a bar chart, and when it encounters None, it throws a TypeError.

#ðŸ’¡ Tip: It's always good to check for None or NaN values before plotting:
#print(df["Product_Category"].isnull().sum())

#1. Option 1: Drop rows where Product_Category is missing
df = df.dropna(subset=["Product_Category"])
#2. Option 2: Replace missing values with a label (e.g. "Unknown")
    #df["Product_Category"] = df["Product_Category"].fillna("Unknown")

#df
labels = df["Product_Category"]
values = df["Revenue"]
plt.bar(df["Product_Category"], df["Revenue"])


#plt.xticks(rotation=45, ha='center')  # Rotate labels 45Â° and align them to the right/center
plt.figure(figsize=(12, 6))
plt.xticks(ticks=range(0, len(labels), 2), labels=labels[::2], rotation=45, ha='right')

plt.tight_layout()  # Automatically adjust layout to prevent clipping

plt.show()
```
